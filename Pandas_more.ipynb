```python

def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Sam Walton'
    return cell

def convert_price_cell(cell):
    if cell=="n.a.":
        return 50
    return cell
    
df = pd.read_excel("DataSet/stock_data.xlsx","Sheet1", converters= {
        'people': convert_people_cell,
        'price': convert_price_cell
    })
df
```

```python
# Export in a sheet in  Excel file
df.to_excel("DataSet/new.xlsx", sheet_name="stocks", index=False, startrow=2, startcol=1)
```

```python
df = pd.read_csv("DataSet/weather_data.csv",parse_dates=['day'])
type(df.day[0])
df
```

```python
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
# ffill is forward fill, fill NaN value by copying previous value index value 
new_df = df.fillna(method="ffill")
new_df
```

```python
# bfill is backward fill, fill NaN value by copying next value index value 
new_df = df.fillna(method="bfill")
new_df
```

```python
# bfill is backward fill, fill NaN value by copying next value index value (column index)
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
# limit 1 means it only fill 1 next value 
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
# Interpolate average previous and next index 
new_df = df.interpolate()
new_df
```

```python
# time interpolate calculate the difference of time and average it 
new_df = df.interpolate(method="time") 
new_df
```


Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)

There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc. Just google "dataframe interpolate" to see complete documentation

```python
new_df = df.dropna()
new_df
```

```python
# how = 'all' drop rows where every column is NaN
new_df = df.dropna(how='all')
new_df
```

```python
# thresh = 1 if there is atleast 1 value (non NaN) in rows it will not drop
new_df = df.dropna(thresh=1)
new_df
```

```python
import numpy as np
df = pd.read_csv("DataSet/weather_data_6.csv")
df
```

```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)
new_df
```

```python
new_df = df.replace({
        -99999: np.nan,
        'no event': 'Sunny',
    })
new_df
```
