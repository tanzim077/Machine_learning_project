### Manipulating Data

**In real life example when we working with a huge amount of data, there are a lot of chance to getting missing data.In order to handle this kind of situation we use some function of pandas. Here we describe about it.**

```python
# Temporary change the value but it doesnt store the value in original file 
car_sales["Make"].str.lower() 
```

```python
car_sales
```

```python
# Methods for change the value but it doesnt store the value in original file 
car_sales["Make"] = car_sales["Make"].str.lower() 
```

### Handle Missing Value

```python
car_sales_missing = pd.read_csv("DataSet/car-sales-missing-data.csv")
```

```python
car_sales_missing
```

To fill the "NaN" value we mean the value of the coloumn

```python
car_sales_missing["Odometer"].mean()
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

Another way to keep the change

```python
car_sales_missing["Doors"].fillna(car_sales_missing["Doors"].mean(), 
                              inplace = True)
```

```python
car_sales_missing
```

```python
car_sales_missing_modify_1 = car_sales_missing
```

#### Export as csv

```python
car_sales_missing_modify_1 = car_sales_missing_modify_1.to_csv("DataSet/car_sales_missing_modify_1.csv", index=False)
```

```python
car_sales_missing_modify_1 = pd.read_csv('DataSet/car_sales_missing_modify_1.csv')
```

```python
car_sales_missing_modify_1
```

```python
import numpy as np
```

```python
# car_sales_missing_modify_1["Price"].str.replace('[\$\,\.]', '').astype(str)
car_sales_missing_modify_1.Price = car_sales_missing_modify_1.Price.replace(np.nan, 0, regex=True)
car_sales_missing_modify_1
```

```python
car_sales_missing_modify_1.Price = car_sales_missing_modify_1["Price"].str.replace('[\$\,\.]', '').astype(float)
```

```python
car_sales_missing_modify_1.Price.fillna(car_sales_missing_modify_1.Price.mean(),inplace=True)
```

```python
car_sales_missing_modify_1
```

### Data Shuffle

```python
# 0-1 for full data. here 1 means 100%
car_sales.sample(frac=1)
```

```python
car_sales_shuffle = car_sales.sample(frac=.8)
```

```python
car_sales_shuffle
```

```python
# Reset Index 
# drop used for drop previous index
car_sales_shuffle.reset_index(drop=True, inplace=True)
```

```python
car_sales_shuffle
```

#### Using Lambda Function

```python
# Conver Km to Miles
car_sales_shuffle["Odometer (KM)"] = car_sales_shuffle["Odometer (KM)"].apply(lambda x: x/1.6 )
```

```python
# Rename Column Name
car_sales_shuffle.rename(columns={"Odometer (KM)":"Odometer (M)"}, inplace=True)
car_sales_shuffle
```
