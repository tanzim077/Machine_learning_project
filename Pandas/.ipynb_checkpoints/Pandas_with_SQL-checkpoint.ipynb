### Read only selected columns

```python
df = pd.read_sql_table('customer', engine, columns = ["Name"])
df
```

### Join two tables and read them in a dataframe using read_sql_query

```python
df = pd.read_sql_query("select id,name from customer",engine)
df
```

```python
query = '''
 SELECT customer.name, customer.city, orders.product_id
 FROM customer INNER JOIN orders
 ON customer.ID = orders.customer_id
'''
df = pd.read_sql_query(query,engine)
df
```

read_sql is a wrapper around read_sql_query and read_sql_table

```python
query = '''
 SELECT product.Name, product.Size, orders.Customer_id
 FROM product INNER JOIN orders
 ON product.ID=orders.product_id
'''
pd.read_sql(query,engine)
```

```python
pd.read_sql("customer",engine)
```

### Write to MSSQL database using to_sql

```python
df = pd.read_csv("../DataSet/customers.csv")
df
```

```python
df.rename(columns={
    'Customer Name': 'Name',
    'Customer Phone': 'Contact'
}, inplace=True)
df
```

```python
df.to_sql(
    name='customer', # database table name
    con=engine,
    if_exists='append',
    index=False
)
```
