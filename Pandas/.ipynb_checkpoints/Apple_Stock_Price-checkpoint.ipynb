## Apple Stock Data 

```python
apple = pd.read_csv("../DataSet/AppleHistoricalQuotes.csv",parse_dates=["Date"], index_col="Date")
```

```python
apple.index
```

```python
apple.columns
```

```python
# Change Column Name
apple.rename(columns={" Close/Last":"Close", " Volume" : "Volume", " Open" : "Open", 
                      " High" : "High", " Low" : "Low"}, inplace=True)
```

```python
apple.columns
```

```python
apple
```

```python
apple["Close"] = apple["Close"].str.replace('$', '')
apple["Open"] = apple["Open"].str.replace('$', '')
apple["High"] = apple["High"].str.replace('$', '')
apple["Low"] = apple["Low"].str.replace('$', '')
```

```python
apple
```

```python
apple["Close"] = apple["Close"].astype(float)
apple["Open"] = apple["Open"].astype(float)
apple["High"] = apple["High"].astype(float)
apple["Low"] = apple["Low"].astype(float)
```

```python
apple.describe
```

```python
apple['2018-01':'2017-01']
```

```python
apple.loc['2017-06'].Close.mean()
```

```python
apple['Close'].resample('M').mean().head(20).plot()
```

```python
apple['Close'].resample('Q').mean().head(20).plot()
```

```python
%matplotlib inline
apple['Close'].plot()
```
